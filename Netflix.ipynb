## Data cleaning

```python
net['rating'].value_counts()
```

```python
net['rating']=net['rating'].replace({'TV-Y7-FV':'TV-Y7'})
net.rating.unique()
```

```python
net['rating']=net['rating'].replace({
    'TV-MA':'adults',
    'TV-PG':'kids with parental guidance',
    'TV-14':'teens age above 14',
    'R':'adults',
    'PG-13':'kids with parental guidance',
    'NR':'NR',
    'PG':'kids with parental guidance',
    'TV-Y7':'kids above age 7',
    'TV-G':'kids',
    'TV-Y':'kids',
    'G':'G',
    'UR':'UR',
    'NC-17':'adults'
})
```

```python
# net['rating'].unique().size
net['rating'].nunique()
```

```python
net['rating'].unique()
```

## missing values

```python
net.country=net.country.fillna(net.country.mode()[0])
```

```python
net.country.isna().sum()
```

```python
net.isnull().sum()
```

```python
net.director.replace(np.nan,'None',inplace=True)
net.cast.replace(np.nan,'None',inplace=True)
net.dropna(inplace=True)
net.drop_duplicates(inplace=True)
net.isna().sum()
```

```python
df_clean=net.drop(['show_id','cast','description'],axis=1)
df_clean.head()
```

### first we have to seperate country before we doing any visualization

```python
# splitting the companies in different 
country_df=df_clean[['type','title','country']]
country_df=(country_df.drop('country',axis=1)
            .join(country_df.country.str.split(',',expand=True).stack().reset_index(drop=True,level=1).rename('country')))
country_df.head()
```

```python
country_df.country.str.split(',',expand=True).stack()
```

```python
top_ten=country_df.country.value_counts().sort_values(ascending=False)[:10]
top_ten
```

```python
x_countries=top_ten.index
y_countries=top_ten.values
plt.figure(figsize=(15,8))
plt.xlabel('country')
plt.ylabel('count')
plt.title('count of shows')
sns.barplot(x=x_countries,y=y_countries,palette='rocket')
plt.show()
```

```python
temp_df=country_df.loc[country_df['country'].isin(top_ten.index)]
temp_df.head()
```

```python
plt.figure(figsize=(17,8))
plot=sns.countplot(data=temp_df,x='type',hue='country',palette='husl')
plot.set_xlabel('countries',weight='bold')
plot.set_ylabel('count of content',weight='bold')
plt.legend(title='country',loc='center right')
plot.figure.suptitle('comparison of top ten content producing countries',y=1,size=20,weight='bold')
```

```python
movie_countries=country_df[country_df['type']=='Movie']
tv_show_countries=country_df[country_df['type']=='TV Show']
```

```python
topten_movie_countries=movie_countries.country.value_counts().sort_values(ascending=False)[:10]
topten_tv_show_countries=tv_show_countries.country.value_counts().sort_values(ascending=False)[:10]
```

```python
plt.subplots(figsize=(15,8))
sns.barplot(topten_movie_countries.index,topten_movie_countries.values,palette='cubehelix')
plt.title('top ten movie producing countries',weight='bold',fontsize=15)
plt.xlabel('countries',weight='bold',fontsize=10)
plt.ylabel('count of movies',weight='bold',fontsize=10)
plt.show()

plt.subplots(figsize=(15,8))
sns.barplot(topten_tv_show_countries.index,topten_tv_show_countries.values)
plt.title('top ten tv show producing countries',weight='bold',fontsize=15)
plt.xlabel('countries',weight='bold',fontsize=10)
plt.ylabel('count of tv show',weight='bold',fontsize=10)
plt.show()
```

### movie based genre

```python
df_temp=df_clean.query("type == 'Movie'")
df_temp=df_temp[['title','type','country','listed_in']]
df_temp
```

```python
df_temp=(df_temp.drop('listed_in',axis=1).join(df_temp.listed_in.str.split(',',expand=True).stack().reset_index(drop=True,level=1).rename('Genre')))
df_temp
```

```python
df_temp.Genre.nunique()
```

```python
genre_df=pd.DataFrame(df_temp.Genre.value_counts().sort_values(ascending=False).reset_index().rename(columns={'index':'genre','Genre':'count'}))
genre_df.head()
```

```python
from wordcloud import WordCloud
from scipy import signal
genres=genre_df.genre.dropna(False)
plt.subplots(figsize=(10,10))
wordcloud=WordCloud(background_color='white',width=500,height=250).generate(' '.join(genres))
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```

```python
plt.subplots(figsize=(20,6))
sns.barplot(genre_df.genre,genre_df.count)
plt.show()
```
